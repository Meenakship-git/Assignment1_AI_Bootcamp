```python
from getpass import getpass
import os
from google import genai

# Ask for API key only once
if "GEMINI_API_KEY" not in os.environ:
    os.environ["GEMINI_API_KEY"] = getpass("Enter your GEMINI_API_KEY: ")

# Create Gemini client
client = genai.Client()
print("âœ… Gemini client ready")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

# Low temperature = more focused and less random
brain_llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.4
)

print("âœ… Brainstorming LLM ready")
```

```python
def search_context(topic: str) -> str:
    """
    'Search tool' â€“ collects useful context, trends and problems
    for the given hackathon theme.
    """
    prompt = f"""
You are a research helper for hackathon teams.

Goal:
- Collect useful context for a hackathon theme.
- Imagine you did some web search.
- Do NOT invent complex data, just give a smart overview.

Follow this structure:

TOPIC:
- Briefly restate the theme.

TRENDS:
- 3â€“5 important current trends related to this theme.

PROBLEMS:
- 3â€“5 real-world problems or pain points.

USERS:
- 3â€“5 types of users or organisations who would care.

Keep it short, clear, and easy to read.

Hackathon theme: {topic}
"""

    resp = brain_llm.invoke(prompt)
    return resp.content
```

```python
def brainstorm_hackathon_ideas(topic: str, context_notes: str) -> str:
    """
    'Brainstorming tool' â€“ uses the context from search_context()
    to create hackathon ideas and choose one best idea with a pitch.
    """
    prompt = f"""
You are helping students come up with hackathon ideas.

Theme: {topic}

Here is some context from a prior 'search' step:
----------------
{context_notes}
----------------

TASK:

1. Based on the context, propose EXACTLY 3 hackathon project ideas:
   - Give each idea a short NAME.
   - 1â€“2 lines description.
   - Mention which users it helps.

2. Then CHOOSE ONE BEST IDEA for a beginner-friendly hackathon project.
   - Clearly mark it as: BEST IDEA.

3. For the BEST IDEA, write a short hackathon PITCH with these headings:

PROBLEM:
- What problem does it solve?

SOLUTION:
- What does the app/agent do?

TECH STACK:
- Simple tools students can use (for example: Python, Colab, Gemini API, simple web UI, etc.)

IMPACT:
- Why is this idea useful / cool for a hackathon?

Keep the language simple and motivating.
"""

    resp = brain_llm.invoke(prompt)
    return resp.content
```

```python
def generate_hackathon_pitch():
    print("ðŸš€ Hackathon Idea Generator Agent")
    print("--------------------------------")

    # 1. Ask user for a theme
    topic = input("Enter a hackathon theme (e.g. 'Smart Personal Finance + AI'): ")

    # 2. Call the Search tool
    print("\n[1/3] Running SEARCH tool to collect context...")
    context_notes = search_context(topic)
    print("\n===== SEARCH CONTEXT =====")
    print(context_notes)

    # 3. Call the Brainstorming tool
    print("\n[2/3] Running BRAINSTORMING tool to generate ideas...")
    ideas_and_pitch = brainstorm_hackathon_ideas(topic, context_notes)

    # 4. Show final pitch
    print("\n===== HACKATHON IDEAS & FINAL PITCH =====")
    print(ideas_and_pitch)

    print("\n[3/3] âœ… Done! You now have a hackathon idea + pitch.")
```

```python
generate_hackathon_pitch()
```
